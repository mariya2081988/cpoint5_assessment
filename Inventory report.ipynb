```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_excel('combined Inventory Dataset.xlsx')
df=pd.DataFrame(data)
df['Date']=pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
weekly_report = df.groupby('Item Number').resample('W-Sat').sum('Inventory')
pivot_table = weekly_report.pivot_table(index='Item Number', columns='Date', values='Inventory', fill_value=0)
pivot_table.to_excel('weekly_inventory_report.xlsx')
monthly_inventory = df.resample('M').sum('Inventory')
monthly_inventory.plot(kind='bar', legend=False)
plt.title('Monthly Inventory Levels')
plt.xlabel('Month')
plt.ylabel('Inventory Level')
plt.show()
plt.savefig('monthly_inventory_bar_chart.png')
```
